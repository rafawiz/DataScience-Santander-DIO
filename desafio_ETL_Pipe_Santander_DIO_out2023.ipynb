```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# URL da API da NASA APOD
api_url = 'https://api.nasa.gov/planetary/apod'

# chave demo para poder usar a API
chave_de_acesso = {
    'api_key': 'DEMO_KEY',
    'hd': True,
}
# solicitação GET na API
response = requests.get(api_url, params=chave_de_acesso)

# se resposta ok
if response.status_code == 200:
    resposta = response.json()

    # obter a imagem astronômica do dia
    img_url = resposta['url']
    titulo = resposta['title']

    # baixar e salvar a imagem
    img_response = requests.get(img_url)
    img_dados = BytesIO(img_response.content)
    imagem = Image.open(img_dados)
    imagem.save('astronomic_image_of_the_day_by_NASA.png')

    # mostrar
    plt.figure(figsize=(8, 8))
    plt.imshow(imagem)
    plt.title(titulo)
    plt.axis('off')
    plt.show()

    print("\nImagem Astronômica do Dia! by NASA")

else:
    print(f"Ocorreu um erro na solicitação à API, status: {response.status_code}")
```
